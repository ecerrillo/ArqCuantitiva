```python
#@title Código de preparación (ejecutar pulsando "play") { run: "auto", vertical-output: true, form-width: "20px", display-mode: "form" }
import matplotlib.pyplot as plt #importa una biblioteca de gráficos
import numpy as np #importa una biblioteca de matrices (no necesitáis saber qué es)
import scipy.stats as stats #importa una biblioteca de estadística

def normal(media,desvest, dib_linea=False, prob=False):
  x = np.linspace(media - 3*desvest, media + 3*desvest, 100)
  plt.figure(figsize=(15,5))
  plt.plot(x, stats.norm.pdf(x, media, desvest))
  plt.fill_between(x, stats.norm.pdf(x, media, desvest), alpha=0.4)
  plt.axvline(media, color='k', linewidth=4)

  if dib_linea == True:
    plt.axvline(media+desvest, color='g', linestyle ='dashed', linewidth=3)
    plt.axvline(media-desvest, color='g', linestyle ='dashed', linewidth=3)
    plt.axvline(media-desvest, color='g', linestyle ='dashed', linewidth=3)
    plt.axvline(media+(desvest*2), color='b', linestyle ='dashed', linewidth=2)
    plt.axvline(media-(desvest*2), color='b', linestyle ='dashed', linewidth=2)
    plt.axvline(media+(desvest*3), color='r', linestyle ='dashed', linewidth=1)
    plt.axvline(media-(desvest*3), color='r', linestyle ='dashed', linewidth=1)

  if prob == True:
    print('Con una probablidad de 1 sigma (68,4%) el intervalo es [{},{}]. Línea verde.'.format(media-desvest,media+desvest))
    print('Con una probablidad de 2 sigma (95,4%) el intervalo es [{},{}]. Línea azul'.format(media-(desvest*2),media+(desvest*2)))
    print('Con una probablidad de 3 sigma (99,7%) el intervalo es [{},{}]. Línea roja'.format(media-(desvest*3),media+(desvest*3)))

  plt.show()

def calcular_prob(media,desvest,int_inf,int_sup):

  if int_inf > int_sup:
    print('ERROR: el intervalo menor no puede ser mayor que el intervalo menor')
  else:
    inf = stats.norm(media, desvest).cdf(int_inf)
    sup = stats.norm(media, desvest).cdf(int_sup)
    x = np.linspace(media - 3*desvest, media + 3*desvest, 100)
    px=np.arange(int_inf,int_sup,0.1)
    plt.figure(figsize=(15,5))
    plt.fill_between(px, stats.norm(media, desvest).pdf(px), alpha=0.4)
    plt.plot(x, stats.norm.pdf(x, media, desvest))
    plt.axvline(media, color='k', linewidth=4)
    
    print('La probabilidad es de {}, es decir un {}%'.format(round(sup-inf,4), round((sup-inf)*100,2)))
```

```python
#@title Probabilidad y distribución normal { run: "auto", vertical-output: true, form-width: "50%", display-mode: "code" }
#@markdown Introducir media y desviación estándar

media = 6033 #@param {type:"number"}
desvest = 40 #@param {type:"number"}
Dibujar_lineas = "Si" #@param ["Si", "No"]
Imprimir_probabilidades = "Si" #@param ["Si", "No"]


if Dibujar_lineas == "Si": 
  trig1 = True
else:
  trig1 = False

if Imprimir_probabilidades == "Si": 
  trig2 = True
else:
  trig2 = False

normal(media,desvest,trig1,trig2)
```

```python
#@title Hallar la probabilidad de un área { run: "auto", vertical-output: true, form-width: "50%", display-mode: "code" }
#@markdown El valor menor del intervalo no deber se mayor al superior

media = 6033 #@param {type:"number"}
desvest = 40 #@param {type:"number"}
int_inf = 5953 #@param {type:"number"}
int_sup = 6113 #@param {type:"number"}

calcular_prob(media,desvest,int_inf,int_sup)
```
